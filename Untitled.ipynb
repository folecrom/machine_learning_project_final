# Différents graphiques pour visualiser les données
Le choix du type de graphique dépend de la nature de vos données et de l'histoire que vous voulez raconter. Voici quelques-uns des graphiques les plus courants et leurs utilisations :

## 1-Graphiques à barres:

Utilisés pour comparer des valeurs catégorielles.
L'axe des x représente les catégories, tandis que l'axe des y représente les valeurs.
Exemple : Comparaison du nombre de ventes par produit :

```python
import matplotlib.pyplot as plt

# Données
produits = ["Produit A", "Produit B", "Produit C"]
ventes = [100, 250, 150]

# Création du graphique
plt.bar(produits, ventes)
plt.xlabel("Produit")
plt.ylabel("Nombre de ventes")
plt.title("Comparaison des ventes par produit")
plt.show()
```

# 2. Graphiques linéaires:

Utilisés pour visualiser des tendances au fil du temps.
L'axe des x représente le temps, tandis que l'axe des y représente les valeurs.
Exemple : Évolution des températures moyennes mensuelles :

```python
import matplotlib.pyplot as plt

# Données
mois = ["Janvier", "Février", "Mars", "Avril", "Mai", "Juin"]
temperatures = [15, 10, 9, 20, 25, 30]

# Création du graphique
plt.plot(mois, temperatures)
plt.xlabel("Mois")
plt.ylabel("Température moyenne (°C)")
plt.title("Évolution des températures moyennes mensuelles")
plt.show()
```

#  3. Diagrammes circulaires:

Utilisés pour montrer la proportion des catégories dans un ensemble de données.
Chaque catégorie est représentée par une part du gâteau, dont la taille est proportionnelle à sa valeur.
Exemple : Répartition des dépenses par type de budget:

```python
import matplotlib.pyplot as plt

# Données
categories = ["Nourriture", "Logement", "Transport", "Loisirs"]
depenses = [30, 40, 20, 10]

# Création du graphique
plt.pie(depenses, labels=categories, autopct="%1.1f%%")
plt.title("Répartition des dépenses par type de budget")
plt.show()
```

# 4. Nuages de points:

Utilisés pour visualiser la relation entre deux variables.
Chaque point du nuage représente une observation, avec sa position sur les axes correspondant aux valeurs des deux variables.
Exemple : Relation entre la taille et le poids des élèves:

```python
import matplotlib.pyplot as plt

# Données
tailles = [170, 165, 180, 175, 168]
poids = [65, 70, 80, 72, 63]

# Création du graphique
plt.scatter(tailles, poids)
plt.xlabel("Taille (cm)")
plt.ylabel("Poids (kg)")
plt.title("Relation entre la taille et le poids des élèves")
plt.show()
```

# 5. Cartes thermiques:

Utilisées pour visualiser des données sur une grille ou une carte.
La couleur de chaque case représente la valeur de la donnée pour cette position.
Exemple : Répartition des températures sur une carte géographique.


```python
import matplotlib.pyplot as plt
import numpy as np

# Données
données = np.array([
    [10, 15, 20],
    [25, 30, 35],
    [40, 45, 50]
])

# Création de la carte thermique
plt.imshow(données, cmap="viridis")
plt.colorbar()
plt.title("Carte thermique des données")
plt.show()
```

En plus de ces graphiques courants, il existe de nombreux autres types de visualisations de données, tels que les graphiques en aires, les graphiques à bulles, les histogrammes et les graphiques en chandeliers. Le choix du graphique le plus approprié dépendra de vos données spécifiques et de l'objectif de votre visualisation.
